```python
basepath = '.'
for fname in os.listdir(basepath):
    path = os.path.join(basepath, fname)
    if os.path.isdir(path):
        # skip directories
        continue
    else:
        print(fname)
```

```python
import re
#note that r'easy' pattern the "r" helps with escape characters inside a string
files = [f for f in os.listdir('.') if re.match(r'easy', f)]  
files

```

```python
#why use the 'r' --you have to escape the escape character!
print([f for f in os.listdir('.') if re.match(r'\.', f)])
print([f for f in os.listdir('.') if re.match('\\.', f)])
```

```python
#Create a function that looks for a blank line indicating the start of email
def strip_head(text):
    tmp=[]
    flag = False
    for i in text:
        if i=='':
            flag = True
        if flag == True:
            tmp.append(i)
    return tmp
        
```

```python
df = []
example = os.listdir("./easy_ham/")
for i in example:
    with open("./easy_ham/"+i,"r") as f:
        data = [i.strip() for i in f]
        s=' ' #joins each line with a space
    df .append(s.join(strip_head(data)))
    
    
          
```

### A little quickie with word vectors
You can use this to parse emails--just DO NOT USE SPACY as we want Naive Bayes for our Classifier.
Also there are more than just the listed formats--try and get as many quickly as you can.

```python
targets = []
data = []
count = 0
not_spamList = os.listdir("./easy_ham/")
spamList = os.listdir("./spam/")
for i in not_spamList:
    with open("./easy_ham/"+i, "r") as f:
        x = email.message_from_file(f)
    mailType = x.get_content_type()
    if mailType == "text/plain":
        data.append(nlp(x.get_payload()).vector)
        targets.append(0)
    elif mailType == "text/html":
        tmp = BeautifulSoup(x.get_payload(), 'html.parser')
        tmp = tmp.text.replace("\n"," ")
        tmp = nlp(tmp)
        data.append(tmp)
        targets.append(0)
    else:
        print(f"msg {i} not parsed")
        print(x.get_content_type())
    if count % 100 == 0:
        print(count)
    count = count + 1
    if count == 1000:
        break
            
    
```

```python
count = 0
spamList = os.listdir("./spam/")
for i in spamList:
    with open("./spam/"+i, "r") as f:
        try:
            x = email.message_from_file(f)
        except:
            print(f"msg {i} not read")
    mailType = x.get_content_type()
    textType = x.get_content_charset()
    if mailType == "text/plain" and textType == 'iso-8859-1':
        data.append(nlp(x.get_payload()).vector)
        targets.append(1)
    elif mailType == "text/html":
        tmp = BeautifulSoup(x.get_payload(), 'html.parser')
        tmp = tmp.text.replace("\n"," ")
        tmp = nlp(tmp)
        data.append(tmp.vector)
        targets.append(1)
    else:
        print(f"msg {i} not parsed")
        print(x.get_content_type())
        print(textType)
    if count % 100 == 0:
        print(count)
    count = count + 1
    if count == 1000:
        break
```

```python
len(data)==len(targets)
```

```python
import pandas as pd
data_vec = pd.DataFrame(data)
```

```python
data_vec
```

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
model = LogisticRegressionCV(cv=5, scoring='accuracy')
```

```python
model.fit(data_vec,targets)
```

```python
model.C_
```

```python
from sklearn.linear_model import LogisticRegression
final_model = LogisticRegression(C=model.C_[0])
```

```python
#professor flex
from sklearn.model_selection import cross_val_score
cross_val_score(final_model,data_vec,targets,cv=5, scoring='accuracy')
```
